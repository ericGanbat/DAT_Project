```python
oneToTwovectorizer=CountVectorizer(ngram_range=(1,2))
twoToTwovectorizer=CountVectorizer(ngram_range=(2,2))
oneToThreevectorizer=CountVectorizer(ngram_range=(1,3))
threeToThreevectorizer=CountVectorizer(ngram_range=(3,3))
```

```python
def train_and_measure(classifier, x, y):
    
    from sklearn import cross_validation
    
    xtrain, xtest, ytrain, ytest = cross_validation.train_test_split(x, y, test_size=0.3, random_state=1234)
    clf = classifier.fit(xtrain, ytrain)
    
    #Print the accuracy on the test and training dataset
    training_accuracy = clf.score(xtrain, ytrain)
    test_accuracy = clf.score(xtest, ytest)
    
    y_true = ytest
    y_pred = clf.predict(xtest)
    meansquarederror=mean_squared_error(y_true,y_pred)
    print "Accuracy on training data: %0.2f" % (training_accuracy)
    print "Accuracy on test data:     %0.2f" % (test_accuracy)
    print "Mean squared error:        %0.2f" % (meansquarederror)
    
    conf = confusion_matrix(y_true, y_pred)
    conf_temp=np.empty([5,5])
    for i in range(5):
        total=conf[i,:].sum(0)
        conf_temp[i,:]=conf[i,:]/float(total)
    print conf[1,:].sum()
    m, n = 5, 5 
    
    #http://stackoverflow.com/questions/20520246/create-heatmap-using-pandas-timeseries 
    df = pd.DataFrame(conf_temp,
                  columns=[1,2,3,4,5],
                  index=[1,2,3,4,5])

    ax = plt.imshow(df, interpolation='nearest', cmap='Oranges').get_axes()
    ax.set_xticks(np.linspace(0, n-1, n))
    ax.set_xticklabels(df.columns)
    ax.set_yticks(np.linspace(0, m-1, m))
    ax.set_yticklabels(df.index)
    ax.grid('off')
    ax.xaxis.tick_top()
    plt.xlabel('Predicted Score')
    plt.ylabel('Actual Score')
    for i in range(m):
        for j in range(n):
            ax.text(j, i, '{:.2f}'.format(df.iget_value(i, j)),
            size='medium', ha='center', va='center',
            path_effects=[patheffects.withSimplePatchShadow(shadow_rgbFace=(1,1,1))])


def classifier(naiveprediction):
    vectorizer=naiveprediction
    x = vectorizer.fit_transform(yelpreview)
    x_back = x.toarray()
    X=pd.DataFrame(x_back, columns=vectorizer.get_feature_names())
    y=yelpstars
    train_and_measure(naive_bayes.MultinomialNB(), X, y)
```

```python

STOP_WORDS=stopwords.words('english')
def removeNumbers(w):
    w=w.lower()
    return w.translate(None,string.digits)
def removePunctuation(w):
    w=w.lower()
    return w.translate(None, string.punctuation)
      
def jointokens(cleantokens):
    return ' '.join(cleantokens)
    
def removeStopwords(tokens):
    cleantokens=[word for word in tokens if word not in STOP_WORDS]
    return cleantokens
def reviewTokenizer(singlereview):
    singlereview=removeNumbers(singlereview)
    singlereview=removePunctuation(singlereview)
    tokens=nltk.word_tokenize(singlereview)
    cleantokens=removeStopwords(tokens)
    if len(cleantokens)>101:
        cleantokens=cleantokens[:100]
    else:
        cleantokens
    return jointokens(cleantokens)
```
